```python
import pandas as pd
from skimpy import clean_columns
```

```python
#Income, PTALs, Health and Disability; General Health, Household income, House Prices, Life Expectancy, Access to Green Spaces 
```

```python
ward_profiles = pd.read_csv("data/ward-profiles-excel-version.csv", encoding="latin")
ward_profiles.head()
```

```python
open_spaces = pd.read_csv("data/public-open-space-nature-ward_access-to-nature.csv", encoding="latin")
open_spaces
```

```python
open_spaces_edit = open_spaces.copy()
open_spaces_edit = open_spaces_edit[['Ward', 'Borough', '% homes with good access to nature', '% homes with deficiency in access to nature']]
```

```python
ward_profiles.columns
#Income, PTALs, Health and Disability; General Health, Household income, House Prices, Life Expectancy, Access to Green Spaces 
```

```python
ward_profiles_edit = ward_profiles.copy()
ward_profiles_edit = ward_profiles_edit[[
    'New code', 
    '% English is First Language of no one in household - 2011',
    'Median House Price (£) - 2014',
    'Median Household income estimate (2012/13)', 
    'Male life expectancy -2009-13',
    'Female life expectancy -2009-13 ',
    '% area that is open space - 2014',
    '% travel by bicycle to work - 2011',
    'Turnout at Mayoral election - 2012',
    'Crime rate - 2014/15',
    '% BAME - 2011',
    '% All Older people aged 65+ - 2015'
]]
ward_profiles_edit
```

```python
ward_profiles_edit = ward_profiles_edit.rename(columns={"New code": "Ward"})
```

```python
# note: maybe we should take out City of London wards 
# investigate the NaN values and if we should drop them
full_data = ward_profiles_edit.merge(open_spaces_edit, on="Ward", how="left")
full_data
```

```python
# moving borough column to front
borough = full_data.pop('Borough')
full_data.insert(1, 'Borough', borough)
full_data
```

```python
# cleaning names

full_data=clean_columns(full_data)

# filter out boroughs
full_data=full_data.drop(range(625,660))

```

```python
# filter out city of london 
full_data=full_data.drop(full_data[full_data['borough'] == 'City of London'].index)

full_data
```

```python
# normalising some columns

mean_female_life_expectancy = full_data['female_life_expectancy_2009_13'].mean()
full_data['female_life_expectancy_normalised'] = full_data['female_life_expectancy_2009_13'] / mean_female_life_expectancy

mean_male_life_expectancy = full_data['male_life_expectancy_2009_13'].mean()
full_data['male_life_expectancy_normalised'] = full_data['male_life_expectancy_2009_13'] / mean_male_life_expectancy

full_data.head()
```

```python
full_data.to_csv('full_data.csv')
```
